```python
import glob
import numpy as np
import matplotlib.pyplot as plt
import librosa
import pandas as pd
```

```python
dirname = 'D:\\Dataset\\CS573100MIR\\Final\\Multi-task Music Dataset\\'
```

```python
def show_dist(counts, title, xticklabel):
    plt.figure(figsize=(15,5))
    idx = np.arange(1,len(xticklabel)+1)
    plt.bar(idx, counts, width=0.3)
    for i,v in enumerate(counts):
        plt.text(i+0.9, v+3, str(v), fontweight='bold')
    plt.xticks(ticks=idx, labels=xticklabel, rotation=90, fontsize=8)
    plt.ylabel('Counts')
    plt.title(title)
    plt.savefig(f'./analysis/{title}.png', transparent=True, dpi=300, bbox_inches = 'tight')
    plt.show()
```

```python
files = glob.glob(dirname+'/mood/*.txt')
moods = []
for file in files:
    with open(file, 'r') as f:
        line = f.read()
        #if 'Bright' in line:
            #print(file)
            #moods.append(line.split())
        moods.append(line.split())
moods = np.concatenate(moods)
v, cnt = np.unique(moods, return_counts=True)
for v_, cnt_ in zip(v, cnt):
    print(f'{v_}: {cnt_}')
```

```python
#title = 'Co-occurence with Bright'
title = 'All Moods'
show_dist(cnt, title, v)
```

```python
files = glob.glob(dirname+'/genre/*.txt')
genres = []
for file in files:
    with open(file, 'r') as f:
        line = f.read()
        if 'Beatss' in line:
            print(file)
    genres.append(line.split())
genres_all = np.concatenate(genres)
v, cnt = np.unique(genres_all, return_counts=True)
for v_, cnt_ in zip(v, cnt):
    print(f'{v_}: {cnt_}')
```

## Music Data

```python
import os
from pydub import AudioSegment
files = glob.glob(dirname+'mp3\\*.mp3')
for file in files:
    out_file = file.replace('mp3', 'wav')
    if not os.path.exists(out_file):
        sound = AudioSegment.from_mp3(file)
        sound.export(out_file, format="wav")
```

```python
def get_timbre(file):
    y, sr = librosa.load(file)
    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)
    centroid = librosa.feature.spectral_centroid(y=y, sr=sr)
    rolloff = librosa.feature.spectral_rolloff(y=y, sr=sr)
    flux = librosa.onset.onset_strength(y=y, sr=sr).reshape(1,-1)
    
    feats = np.vstack((mfcc, centroid, rolloff, flux))
    mean = np.mean(feats, axis=1)
    std = np.std(feats, axis=1)
    
    t = librosa.frames_to_time(feats.shape[1], sr=sr)
    stds = []
    for i in range(int(t)):
        length = feats.shape[1]//t
        start_idx = int(i*length)
        end_idx = int(min((i+1)*length, feats.shape[1]))
        stds.append(np.std(feats[:,start_idx:end_idx], axis=1))
    stds = np.vstack(stds).T
    mean_std = np.mean(stds, axis=1)
    std_std = np.std(stds, axis=1)
    
    extract_feats = np.hstack((mean, std, mean_std, std_std))
    
    return extract_feats
    
```

```python
ds = pd.DataFrame(files, columns=['path'])
ds['filename'] = ds['path'].apply(lambda x: x.split('\\')[-1])
```

```python
ds['features'] = ds['path'].apply(lambda x: get_timbre(x))
```

```python
out_ds = ds.loc[:,['filename','features']].copy()
out_ds.to_csv(dirname+'timbre_feats.csv', index=False)
```
